```python
planilha = pd.read_excel("/content/Planilha para python.xlsx")
display(planilha)
```

```python
media_idade = planilha["IDADE"].mean()
desvio_idade = planilha["IDADE"].std()
print(f"Mdia da idade: {media_idade:.2f}")
print(f"Desvio padro da idade: {desvio_idade:.2f}\n")

media_peso = planilha["PESO"].mean()
desvio_peso = planilha["PESO"].std()
print(f"Mdia do peso: {media_peso:.2f} kg")
print(f"Desvio padro do peso: {desvio_peso:.2f} kg\n")

media_altura = planilha["ALTURA"].mean()
desvio_altura = planilha["ALTURA"].std()
print(f"Mdia da altura: {media_altura:.2f} m")
print(f"Desvio padro da altura: {desvio_altura:.2f} m\n")
```
